# WikiLinks

```python
url = 'https://en.wikipedia.org/wiki/Python'
```

```python
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
soup.select("#mw-content-text > div.mw-parser-output > ul > li > a")[0]['href']
```

```python
soup.select('#mw-content-text > div.mw-parser-output > ul > li > ul > li > a')[1]['href']
```

```python
wikilinks = []

for i in range(len(soup.select('#mw-content-text > div.mw-parser-output > ul > li > a'))):
    wikilinks.append('wikipedia.org' + soup.select('#mw-content-text > div.mw-parser-output > ul > li > a')[i]['href'])
for i in range(len(soup.select('#mw-content-text > div.mw-parser-output > ul > li > ul > li > a'))):
    wikilinks.append('wikipedia.org' + soup.select('#mw-content-text > div.mw-parser-output > ul > li > ul > li > a')[i]['href'])
wikilinks
```

# Most Wanted

```python
url = 'https://www.fbi.gov/wanted/topten'
response = requests.get(url)
response.status_code
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select('#query-results-0f737222c5054a81a120bce207b0446a > ul > li > h3 > a')
```

```python
mostwanted = []
for i in range(len(soup.select('#query-results-0f737222c5054a81a120bce207b0446a > ul > li > h3 > a'))):
    mostwanted.append(soup.select('#query-results-0f737222c5054a81a120bce207b0446a > ul > li > h3 > a')[i].get_text())

mostwanted
```
